```python
# remove old binary label
data = data.drop(
    columns=['Label'],
    errors='ignore'
)

# features
X = data.drop(
    columns=['Attack Type']
)

# target
y = data['Attack Type']
```

```python
X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.median())
```

```python
X = X.loc[:, X.var() != 0]
```

```python
import joblib
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import (
    LabelEncoder,
    PowerTransformer,
    RobustScaler
)

from sklearn.pipeline import Pipeline

from sklearn.metrics import (

    accuracy_score,

    classification_report,

    confusion_matrix,

    f1_score
)

from xgboost import XGBClassifier

# ==========================================
# LABEL ENCODING
# ==========================================

encoder = LabelEncoder()

y_encoded = encoder.fit_transform(y)

# ==========================================
# TRAIN TEST SPLIT
# ==========================================

X_train, X_test, y_train, y_test = train_test_split(

    X,
    y_encoded,

    test_size=0.2,

    stratify=y_encoded,

    random_state=42
)

# ==========================================
# NUMBER OF CLASSES
# ==========================================

num_classes = len(
    np.unique(y_encoded)
)

print("Number of Classes:",
      num_classes)

print("Classes:",
      encoder.classes_)

# ==========================================
# PIPELINE
# ==========================================

model_pipeline = Pipeline([

    (
        'power',

        PowerTransformer(
            method='yeo-johnson'
        )
    ),

    (
        'scaler',

        RobustScaler()
    ),

    (
        'classifier',

        XGBClassifier(

            # multiclass
            objective='multi:softprob',

            num_class=num_classes,

            # tree params
            max_depth=6,

            learning_rate=0.1,

            n_estimators=300,

            # regularization
            reg_lambda=1,

            reg_alpha=0.5,

            # anti overfitting
            subsample=0.8,

            colsample_bytree=0.8,

            # evaluation
            eval_metric='mlogloss',

            # faster training
            tree_method='hist',

            # cpu usage
            n_jobs=-1,

            random_state=42
        )
    )
])

# ==========================================
# TRAIN MODEL
# ==========================================

print("\nTraining Started...\n")

model_pipeline.fit(
    X_train,
    y_train
)

print("\nTraining Completed\n")

# ==========================================
# PREDICTIONS
# ==========================================

y_pred = model_pipeline.predict(
    X_test
)

# ==========================================
# EVALUATION
# ==========================================

print(
    "Accuracy:",
    accuracy_score(
        y_test,
        y_pred
    )
)

print(
    "\nWeighted F1 Score:",
    f1_score(
        y_test,
        y_pred,
        average='weighted'
    )
)

print(
    "\nClassification Report:\n"
)

print(

    classification_report(

        y_test,

        y_pred,

        target_names=encoder.classes_
    )

)

print(
    "\nConfusion Matrix:\n"
)

print(
    confusion_matrix(
        y_test,
        y_pred
    )
)

# ==========================================
# SAVE PIPELINE
# ==========================================

joblib.dump(

    model_pipeline,

    "ids_multiclass_pipeline1.pkl"
)

# ==========================================
# SAVE ENCODER
# ==========================================

joblib.dump(

    encoder,

    "label_encoder1.pkl"
)

# ==========================================
# SAVE FEATURE COLUMNS
# ==========================================

joblib.dump(

    X.columns.tolist(),

    "feature_columns1.pkl"
)

print(
    "\nAll Files Saved Successfully"
)
```

```python
import joblib
import pandas as pd

# ==========================================
# LOAD MODEL
# ==========================================

model = joblib.load(
    "ids_multiclass_pipeline1.pkl"
)

# ==========================================
# LOAD ENCODER
# ==========================================

encoder = joblib.load(
    "label_encoder1.pkl"
)

# ==========================================
# LOAD FEATURE COLUMNS
# ==========================================

feature_columns = joblib.load(
    "feature_columns1.pkl"
)

# ==========================================
# TAKE REAL DDoS SAMPLE
# ==========================================

sample = data[
    data["Attack Type"] == "DDoS"
].iloc[[0]]

# ==========================================
# REMOVE TARGET
# ==========================================

X_sample = sample.drop(
    columns=["Attack Type"]
)

# ==========================================
# KEEP ONLY TRAINED FEATURES
# ==========================================

X_sample = X_sample[
    feature_columns
]

# ==========================================
# PREDICT
# ==========================================

prediction = model.predict(
    X_sample
)

# ==========================================
# PROBABILITIES
# ==========================================

probabilities = model.predict_proba(
    X_sample
)

# ==========================================
# DECODE LABEL
# ==========================================

predicted_label = (
    encoder.inverse_transform(
        prediction
    )
)

# ==========================================
# RESULTS
# ==========================================

print(
    "Actual Label:",
    sample["Attack Type"].values[0]
)

print(
    "Predicted Label:",
    predicted_label[0]
)

print("\nProbabilities:\n")

for cls, prob in zip(

    encoder.classes_,

    probabilities[0]
):

    print(
        f"{cls}: {prob:.4f}"
    )
```
